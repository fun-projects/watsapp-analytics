```python
import sys
from time import time
from datetime import datetime
import codecs
from pprint import pprint
data=[]

```

```python
def readdata():
    filename="WhatsApp Chat with  CSE WARRIORS  .txt"
    infile = open(filename,'r')
    arq = codecs.open(filename, "r", "utf-8-sig")
    content = arq.read()
    arq.close()
    lines = content.split("\n")
    lines = [l for l in lines if len(l) != 1]
    count=0
    message=""
    entry={}
    for line in lines:
        #print line
        raw=line
        #print raw
        
        if " - " in raw and ": " in raw:
            entry["message"]=message
            data.insert(count,entry)
            count+=1
            entry={}
            message=""
            msg_date, sep, msg_raw = raw.partition(" - ")
            sender, sep,msg  = msg_raw.partition(": ")
            date, sep, raw_time = msg_date.partition(", ")
            #print msg
            datime=date+" "+raw_time
            dt = datetime.strptime(datime, "%d/%m/%Y %I:%M %p")
            entry["datetime"]=dt
            entry["sender"]=sender
            
            message+=msg
            
        else:
            message+=raw
    pprint(data) 
            
readdata()
```
